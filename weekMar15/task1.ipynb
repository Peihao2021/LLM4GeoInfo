```python
cities = cities[['city_st', 'lat', 'lng', 'population', 'lat_prompt', 'lng_prompt', 'distance_km']]
cities = cities[cities['population'] >= 50000]
cities
```

```python
import re
results = pd.read_csv("results.csv")
def standardize_line(line):
    if line.count(',') == 5:
        standardized_line = line
    else:
        parts = re.split(r'(\b[A-Z][a-z]*\b)', line)
        parts = [part.strip() for part in parts if part.strip()]
        standardized_parts = []
        for i in range(0, len(parts), 2):
            city = parts[i]
            if i + 1 < len(parts):
                state = parts[i + 1]
                standardized_parts.append(f"{city}, {state}")
        standardized_line = ', '.join(standardized_parts)
    
    return standardized_line

results['end'] = results['line'].apply(standardize_line)

def replace_double_commas(line):
    cleaned_line = line.replace(',,', ',').strip()
    cleaned_line = ', '.join([part.strip() for part in cleaned_line.split(',')])
    return cleaned_line

results['end'] = results['end'].apply(replace_double_commas)
```

```python
results
```

```python
def split_into_columns(line):
    parts = line.split(', ')
    result = {}
    for i in range(0, len(parts), 2):
        city_key = f"city{i//2 + 1}"
        state_key = f"state{i//2 + 1}"
        result[city_key] = parts[i] if i < len(parts) else ""
        result[state_key] = parts[i+1] if i+1 < len(parts) else ""
    return pd.Series(result)

expanded_columns = results['end'].apply(split_into_columns)
results = pd.concat([results, expanded_columns], axis=1)
```
