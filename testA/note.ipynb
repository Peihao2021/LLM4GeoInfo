```python
import pandas as pd
df = pd.read_csv('uscities.csv')
df.head()
```

```python
df_over_100k = df[df['population'] > 100000].head(160)
df_under_100k = df[df['population'] <= 100000].head(160)
```

```python
import numpy as np
def generate_random_triplets(df, num_triplets=160):
    columns = ['City1', 'City1_state', 'City1_long', 'City1_lat',
               'City2', 'City2_state', 'City2_long', 'City2_lat',
               'City3', 'City3_state', 'City3_long', 'City3_lat']
    data = {col: [] for col in columns}
    
    for _ in range(num_triplets):
        samples = df.sample(n=3)
        
        for i, (idx, row) in enumerate(samples.iterrows()):
            city_prefix = f'City{i+1}'
            data[f'{city_prefix}'].append(row['city'])
            data[f'{city_prefix}_state'].append(row['state_id'])
            data[f'{city_prefix}_long'].append(row['lng'])
            data[f'{city_prefix}_lat'].append(row['lat'])
            
    triplet_df = pd.DataFrame(data)
    return triplet_df

triplets_over_100k = generate_random_triplets(df_over_100k)
triplets_under_100k = generate_random_triplets(df_under_100k)
```

```python
from geopy.distance import geodesic

def calculate_distances(row):
    city1_coords = (row['City1_lat'], row['City1_long'])
    city2_coords = (row['City2_lat'], row['City2_long'])
    city3_coords = (row['City3_lat'], row['City3_long'])

    distance_to_city2 = geodesic(city1_coords, city2_coords).kilometers
    distance_to_city3 = geodesic(city1_coords, city3_coords).kilometers

    closer_city = row['City2'] if distance_to_city2 < distance_to_city3 else row['City3']

    return closer_city

triplets_under_100k_prompted['Closer_City'] = triplets_under_100k_prompted.apply(calculate_distances, axis=1)
triplets_under_100k_prompted

def cor(row):
    gpt_ans = row['Answer']
    cor_ans = row['Closer_City']
    return gpt_ans == cor_ans

triplets_under_100k_prompted['Correct'] = triplets_under_100k_prompted.apply(cor, axis=1)
triplets_under_100k_prompted

sum = 0
for _ , row in triplets_under_100k_prompted.iterrows():
    if row['Correct'] == True:
        sum += 1
print(sum / 160)
```

```python
'''Now, We try ensure using > 500 km, as we did not obtain great results'''
from geopy.distance import great_circle

def calculate_distance(lat1, lon1, lat2, lon2):
    return great_circle((lat1, lon1), (lat2, lon2)).km

def generate_random_triplets(triplets_over_100k, number_of_triplets=160):
    triplets_data = {
        "City1": [], "City1_state": [], "City1_lat": [], "City1_long": [],
        "City2": [], "City2_state": [], "City2_lat": [], "City2_long": [],
        "City3": [], "City3_state": [], "City3_lat": [], "City3_long": [],
        "Valid_distance_difference": []
    }
    for _ in range(number_of_triplets):
        valid_triplet_found = False
        attempts = 0
        while not valid_triplet_found and attempts < 100:
            cities_sample = triplets_over_100k.sample(3).reset_index(drop=True)
            city1, city2, city3 = cities_sample.iloc[0], cities_sample.iloc[1], cities_sample.iloc[2]
            distance1_2 = calculate_distance(city1['lat'], city1['lng'], city2['lat'], city2['lng'])
            distance1_3 = calculate_distance(city1['lat'], city1['lng'], city3['lat'], city3['lng'])
            
            if np.abs(distance1_2 - distance1_3) > 500:
                for i, city in enumerate([city1, city2, city3], start=1):
                    triplets_data[f"City{i}"].append(city['city'])
                    triplets_data[f"City{i}_state"].append(city['state_id'])
                    triplets_data[f"City{i}_lat"].append(city['lat'])
                    triplets_data[f"City{i}_long"].append(city['lng'])
                
                triplets_data["Valid_distance_difference"].append(np.abs(distance1_2 - distance1_3) > 500)
                valid_triplet_found = True    
            attempts += 1
            
    return pd.DataFrame(triplets_data)
triplets_df_over_500 = generate_random_triplets(df)
```
