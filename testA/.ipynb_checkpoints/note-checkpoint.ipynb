```python
import pandas as pd
df = pd.read_csv('uscities.csv')
df.head()
```

```python
df_over_100k = df[df['population'] > 100000].head(160)
df_under_100k = df[df['population'] <= 100000].head(160)
```

```python
df_over_100k.head()
```

```python
df_under_100k.head()
```

```python
import numpy as np
def generate_random_triplets(df, num_triplets=160):
    columns = ['City1', 'City1_state', 'City1_long', 'City1_lat',
               'City2', 'City2_state', 'City2_long', 'City2_lat',
               'City3', 'City3_state', 'City3_long', 'City3_lat']
    data = {col: [] for col in columns}
    
    for _ in range(num_triplets):
        samples = df.sample(n=3)
        
        for i, (idx, row) in enumerate(samples.iterrows()):
            city_prefix = f'City{i+1}'
            data[f'{city_prefix}'].append(row['city'])
            data[f'{city_prefix}_state'].append(row['state_id'])
            data[f'{city_prefix}_long'].append(row['lng'])
            data[f'{city_prefix}_lat'].append(row['lat'])
            
    triplet_df = pd.DataFrame(data)
    return triplet_df

triplets_over_100k = generate_random_triplets(df_over_100k)
triplets_under_100k = generate_random_triplets(df_under_100k)
```

```python
gpt_over_100k = triplets_over_100k[["City1", "City1_state", "City2", "City2_state", "City3", "City3_state"]]
```

```python
pd.options.display.max_rows = 15
gpt_over_100k
```

```python
prompts = gpt_over_100k.apply(lambda row: f"Is {row['City1']}, {row['City1_state']} closer to {row['City2']}, {row['City2_state']} or {row['City3']}, {row['City3_state']}?", axis=1)
gpt_over_100k_prompts = pd.DataFrame(prompts, columns=['Prompt'])
gpt_over_100k_prompts
gpt_over_100k_prompts.to_csv("over_100k_prompts.csv")
```

```python
gpt_under_100k = triplets_under_100k[["City1", "City1_state", "City2", "City2_state", "City3", "City3_state"]]
gpt_under_100k
prompts = gpt_under_100k.apply(lambda row: f"Is {row['City1']}, {row['City1_state']} closer to {row['City2']}, {row['City2_state']} or {row['City3']}, {row['City3_state']}?", axis=1)
gpt_under_100k_prompts = pd.DataFrame(prompts, columns=['Prompt'])
gpt_under_100k_prompts
gpt_under_100k_prompts.to_csv("under_100k_prompts.csv")
```

```python
'''Note at this point, ChatGPT has populated the response onto the spreadsheet using an internal software. '''
triplets_over_100k_prompted = pd.read_csv('triplets_over_100k_prompted.csv')
triplets_over_100k_prompted = triplets_over_100k_prompted.drop("Unnamed: 0", axis=1)
triplets_over_100k_prompted.head()
```

```python
'''Note at this point, ChatGPT has populated the response onto the spreadsheet using an internal software. '''
triplets_under_100k_prompted = pd.read_csv('triplets_under_100k_prompted.csv')
triplets_under_100k_prompted = triplets_under_100k_prompted.drop("Unnamed: 0", axis=1)
triplets_under_100k_prompted.head()
```

```python
from geopy.distance import geodesic

def calculate_distances(row):
    city1_coords = (row['City1_lat'], row['City1_long'])
    city2_coords = (row['City2_lat'], row['City2_long'])
    city3_coords = (row['City3_lat'], row['City3_long'])

    distance_to_city2 = geodesic(city1_coords, city2_coords).kilometers
    distance_to_city3 = geodesic(city1_coords, city3_coords).kilometers

    closer_city = row['City2'] if distance_to_city2 < distance_to_city3 else row['City3']

    return closer_city

triplets_under_100k_prompted['Closer_City'] = triplets_under_100k_prompted.apply(calculate_distances, axis=1)
triplets_under_100k_prompted

def cor(row):
    gpt_ans = row['Answer']
    cor_ans = row['Closer_City']
    return gpt_ans == cor_ans

triplets_under_100k_prompted['Correct'] = triplets_under_100k_prompted.apply(cor, axis=1)
triplets_under_100k_prompted

sum = 0
for _ , row in triplets_under_100k_prompted.iterrows():
    if row['Correct'] == True:
        sum += 1
print(sum / 160)
```
